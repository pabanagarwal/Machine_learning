```python
# import the necessary packages
import numpy as np
import urllib3
import cv2
import os
import json
import sys
from pprint import pprint
from PIL import Image

SRC_JSON_FILE_LOC="C:\\GreatLakes\\Capstone\\Data\\ImageDataJSON\\"
TGT_IMG_FILE_LOC="C:\\GreatLakes\\Capstone\\Data\\"

if not os.path.exists(TGT_IMG_FILE_LOC+"train\\RGB\\"):
    os.makedirs(TGT_IMG_FILE_LOC+"train\\RGB\\")
    
TRAIN_JSON_FILE="train.json"

with open(SRC_JSON_FILE_LOC+TRAIN_JSON_FILE) as data_file:    
    ImgTrainData = json.load(data_file)
#pprint(ImgTrainData)
```

```python
http = urllib3.PoolManager()
urllib3.disable_warnings()
resp = urllib3.HTTPSConnectionPool('localhost', port=9021, cert_reqs='CERT_NONE',assert_hostname=False,timeout=1)
```

```python
# Loading TRAIN image data
images_skipped=0
for im in ImgTrainData['images']:
    if int(im['image_id']) < 5120:
        continue
    try:
        list_var = im['url'][0].split('/')
        name=list_var[len(list_var)-1]
        ID_name='ID_'+str(im['image_id'])+'_NAME_'+name
        print(ID_name)
        filename_part = ID_name.split('.')
        if len(filename_part)>1:
            if   filename_part[1]=='SS2':
                ID_name=filename_part[0]+'.jpg'
        else:
            ID_name=filename_part[0]+'.jpg'
            
        resp = http.request('GET', im['url'][0])
        image = np.asarray(bytearray(resp.data), dtype="uint8")
        if str(type(image))=='NoneType':
            print("       Img Skipped " + im['url'][0])
            images_skipped=images_skipped+1
            continue

        image_clr = cv2.imdecode(image, cv2.IMREAD_COLOR)

        if str(type(image_clr))=="<class 'NoneType'>":
            print("       Img Skipped " + im['url'][0])
            images_skipped=images_skipped+1
            continue

        ID_name_rgb='RGB_' + ID_name

        cv2.imwrite(TGT_IMG_FILE_LOC+"train\\RGB\\"+ ID_name_rgb,image_clr)
    except:
        print("       Img Skipped " + im['url'][0])
        images_skipped=images_skipped+1
        continue
print ("\n\nTotal images skipped: %d " % images_skipped)
```
