## 1. Predict the image

```python
import numpy as np
import cv2
```

```python
### Load entire model

from keras.models import load_model

# Returns a compiled model identical to the previous one
model_2 = load_model('C:\\MY_Personal\\GREATLAKES\\CAPSTON\\my_own\\vgg16_final.h5')
```

```python
print(model_2.metrics_names)
```

```python
input_image='C:\\MY_Personal\\GREATLAKES\\CAPSTON\\my_own\\sofa.jpg'
```

```python
def image_predict(input_image):
    inp_image_read=cv2.imread(input_image)
    inp_img_arr = np.array(inp_image_read)
    inp_img = cv2.resize(inp_img_arr, (150,150) )
    img_2D_data_all = np.array([np.array(inp_img) ],'f')
    x_validate = img_2D_data_all/255
    
    return model_2.predict(x_validate)
    
```

```python
output=image_predict(input_image)
```

## 2. Object detection and bounding box

```python
# object detection example
# usage: python object_detection.py <input_image>

# import necessary packages
import cvlib as cv
from cvlib.object_detection import draw_bbox
import sys
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read input image
# C:\\MY_Personal\\GREATLAKES\\CAPSTON\\my_own\\all.jpg
# C:\\MY_Personal\\GREATLAKES\\CAPSTON\\train\\GR_SML\\image_id_16__575d3181Na7521aa3.jpg.jpg
image = cv2.imread('C:\\MY_Personal\\GREATLAKES\\CAPSTON\\train\\images\\image_id_23__58675f96N6c9eadc4.jpg')
```

```python
# apply object detection
bbox, label, conf = cv.detect_common_objects(image)
print(bbox, label, conf)
```

```python
# draw bounding box over detected objects
out = draw_bbox(image, bbox, label, conf)

# display output
# press any key to close window           
##cv2.imshow("object_detection", out)
##cv2.waitKey()

# save output
##cv2.imwrite("object_detection.jpg", out)

cv2.imwrite('C:\\MY_Personal\\GREATLAKES\\CAPSTON\\my_own\\object_detection.jpg',out)

# release resources
##cv2.destroyAllWindows()
```
